```python
import zipfile
#Descomprimir archivo
with zipfile.ZipFile('bbc-news-data.csv.zip', 'r') as f:
    f.extractall()
```

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
#Leer datos
df = pd.read_csv('bbc-news-data.csv', delimiter='\t')

content = df['content'].to_list() #sentences

tokenizer = Tokenizer()

#Generar diccionario de word_index
tokenizer.fit_on_texts(content)
word_index = tokenizer.word_index

#Generar las secuencias de numeros para cada oracion
sequences = tokenizer.texts_to_sequences(content)
```

```python
#Dar uniformidad a las matrices de secuencias para entrar a la red
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded = pad_sequences(sequences)
print(padded[:5])
print(padded.shape)
```
